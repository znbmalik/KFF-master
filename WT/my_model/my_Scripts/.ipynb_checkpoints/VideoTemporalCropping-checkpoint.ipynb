```python
import os
import numpy as np
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
from moviepy.video.io.VideoFileClip import VideoFileClip
```

```python
path2 = '../my_data/Test_Input/'
my_file = open("../my_data/LTA.txt", "r")
classes = []
for line in my_file:
    stripped_line = line.strip()
    classes.append(stripped_line)

my_file.close()

print(classes)
```

```python
#classes=['BabyCrawling','BlowingCandles','BodyWeightSquats','HandStandPushups','HandstandWalking','JumpingJack','Lunges','PullUps','PushUps','RockClimbingIndoor','RopeClimbing','Swing','TaiChi','TrampolineJumping','WalkingWithDog','WallPushups']
duratn=[]

for cls in classes:
    for filename in os.listdir(path2+cls):
        folder=path2+cls
        filename=os.path.join(folder,filename)
        #print(filename)
        clip = VideoFileClip(filename) # <= !!!
        #print(clip.duration)
        duratn.append(clip.duration)
        
smaller_clip=np.min(np.array(duratn))
print(smaller_clip)
```

```python
def iterate_subclip(vid,output_video):
    video = VideoFileClip(vid)
    subclip = video.subclip(0, 1)
    fps=30
    subclip.write_videofile(output_video, fps=fps, codec='png') 
    
    #del subclip
    subclip.close()
    video.close()
    #del video
```

```python

path1 = '../my_data/Test_Input/'
path2= '../my_data/Crop_Videos/'
#classes=['BabyCrawling','BlowingCandles','BodyWeightSquats','HandStandPushups','HandstandWalking','JumpingJack','Lunges','PullUps','PushUps','RockClimbingIndoor','RopeClimbing','Swing','TaiChi','TrampolineJumping','WalkingWithDog','WallPushups']
for cls in classes:
    subpathInput=path1+cls
    subpathOutput=path2+cls
    
    if not os.path.exists(subpathOutput):
        os.makedirs(subpathOutput)
    else:
        files = os.listdir(subpathOutput)
        for f in files:
            os.remove(os.path.join(subpathOutput,f))
            
    for filename in os.listdir(subpathInput):
        count=1
        startTime=0.12#
        endTime=0.84#0.9#
        while count<=1:# here 3 specifying number of clips
            video=os.path.join(subpathInput,filename)
            #output_video=subpathOutput+'/clip_'+str(count)+'_'+filename
            output_video=subpathOutput+'/'+filename
            ffmpeg_extract_subclip(video, startTime, endTime, targetname=output_video)
            #iterate_subclip(video,output_video)
            startTime+=0.08
            endTime+=0.08
            count+=1
               
            
```
