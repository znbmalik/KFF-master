```python

input_path = os.path.join('/home/my_model/my_data', 'UCF-101')
output_path = os.path.join('/home/my_model/my_data', 'Key-UCF101')
```

```python
def createDirectories(processed_vid):           
    if not os.path.exists(processed_vid):
        os.makedirs(processed_vid)
    else:
        files = os.listdir(processed_vid)
        for f in files:
            os.remove(os.path.join(processed_vid,f))
```

```python
for file in os.listdir(input_path):
    
    #print(file)
    for vid in os.listdir(os.path.join(input_path,file)):
        #print(vid)
    #vid=file.split('.')[0]
        folder_path="/home/my_model/my_data/Key-UCF101/"+str(file)+"/"+str(vid.split('.')[0])
        createDirectories(folder_path)
    #print(folder_path)
```

```python
def extract_frames(input_video_path,output_frame_path):

    
    vidcap = cv2.VideoCapture(input_video_path)
    im_name =  input_video_path.split('/')[-1].split('.')[-2]
    count=1
    
    while(count<=45):
        success ,image1 = vidcap.read()
        print(os.path.join(output_frame_path, im_name+"_"+str(count)+".jpg"))
        cv2.imwrite(os.path.join(output_frame_path, im_name+"_"+str(count)+".jpg"), image1)     # save frame as JPEG file

        
        count=count+1
    
    return count

```

```python

for folder in os.listdir(input_path):
    
    #Will iterate for all folders containing videos or all categories of videos in data set'''
    
    if folder is not '.ipynb_checkpoints':
        fullPath = os.path.join(input_path, folder)
        outfullPath = os.path.join(output_path, folder)
        if os.path.isdir(fullPath):
            for files in os.listdir(fullPath):
                invideo=os.path.join(fullPath,files)
                outvideo=os.path.join(outfullPath,files)
                print(invideo)
                print(outvideo)
                totalFrames=extract_frames(invideo,outvideo.split('.')[0]) #-----------Zainab model
                print("-------------Extracted "+str(totalFrames)+" from "+files+"--------------")
```
