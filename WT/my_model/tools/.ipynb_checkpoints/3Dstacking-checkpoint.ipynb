```python
import flow_vis
import os
import numpy as np
import filecmp
import matplotlib.pyplot as plt
import PIL
from PIL import ImageFilter
from PIL import Image
```

```python
interval=1
depth=2
#LK_Flow_StackedFrames, Brox_Flow_StackedFrames,PLK_Flow_StackedFrames,Tvl1_Flow_StackedFrames,Farneback_Flow_StackedFrames
algo='Brox_Flow_StackedFrames'
list_Npy=['_UP_ROI_Train_Npy','_UP_ROI_Val_Npy','_UP_ROI_Test_Npy']
list_3D_Npy=['_3D_UP_Training_Npy','_3D_UP_Validation_Npy','_3D_UP_Testing_Npy']

#list_UP_ROI_Npy=['_UP_ROI_Train_Npy','_UP_ROI_Val_Npy','_UP_ROI_Test_Npy']
#list_out_UP_ROI_Npy=['_out_UP_ROI_Train_Npy','_out_UP_ROI_Val_Npy','_out_UP_ROI_Test_Npy']

for _Npy,_3D_Npy in zip(list_Npy,list_3D_Npy):
    inputPath='/home/tf-faster-rcnn/my_data/'+algo+'/depth_'+str(depth)+'/'+str(interval)+_Npy
    outputPath='/home/tf-faster-rcnn/my_data/'+algo+'/depth_'+str(depth)+'/'+str(interval)+_3D_Npy
    #print(_UP_ROI_Npy)
    
    if not os.path.exists(outputPath):
        os.makedirs(outputPath)
    else:
        for f in os.listdir(outputPath):
            if f is not ".ipynb_checkpoints":
                os.remove(os.path.join(outputPath,f))
    
    
    for file in os.listdir(inputPath):
        npy_data1=np.load(os.path.join(inputPath,file))
        framelist=[]
        i=0

        while i<44:
            a = np.array(npy_data1[i])

            b = np.array(npy_data1[i+1])

            flow_uv=np.stack((a, b)) 
            #print(flow_uv.shape)
            flow_uv=flow_uv.transpose(1,2,0)
            #print(flow_uv.shape)
            flow_color = flow_vis.flow_to_color(flow_uv, convert_to_bgr=False)
            #img = Image.fromarray(flow_color)
            radius = 3
            PIL_image = Image.fromarray(flow_color, 'RGB')
            #img = PIL_image.filter(ImageFilter.GaussianBlur(radius=radius))
            out_path=os.path.join(outputPath,file)
            #print(np.array(PIL_image).shape)

            #img1=np.stack((np.array(img1),np.array(PIL_image)))
            framelist.append(np.array(PIL_image))
            #img.save(out_path)
            i+=2
        frames=np.array(framelist)
        frames=frames.transpose(2,1,0,3)
        print(frames.shape)
        #break
        np.save(out_path,frames)
```
