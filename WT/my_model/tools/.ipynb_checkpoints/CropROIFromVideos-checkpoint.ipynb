```python
import cv2
import os
import numpy as np
```

```python
depth=2
interval=1
input_path = os.path.join('/home/my_model/my_data', 'Crop_Videos')

stacked_path='/home/my_model/my_data/depth_'+str(depth)+'/'+str(interval)+'/'

if not os.path.exists(stacked_path):
    os.makedirs(stacked_path)
else:
    files = os.listdir(stacked_path)
    for f in files:
        os.remove(os.path.join(stacked_path,f))

rpn_path=os.path.join('/home/my_model/my_data', 'RpnResult/')
rpn = open('/home/my_model/my_data/RpnResult/Rpn.txt',"r")
rpn_lines = rpn.readlines()
rpn.close()
```

```python
def check_existing(key_arr, List):
        flag=True
        for index, sublist in enumerate(List):
            if key_arr in sublist:
                flag = False
        return flag
```

```python
def get_rpn(all_rpns,vid_label):
        List = []
        a = 0
        #print(all_rpns)
        while (a < len(all_rpns)/17):
            
            rpn_List = []
            if check_existing(all_rpns[a],List):
                ra = all_rpns[a]
                rpn_List.append(ra)
                b = a + 1
                while (b < len(all_rpns)):

                    rb = all_rpns[b]
                    # overlap between A and B
                    SA = (ra[2] - ra[0]) * (ra[3] - ra[1])  # A['w']*A['h']
                    SB = (rb[2] - rb[0]) * (rb[3] - rb[1]) # B['w']*B['h']
                    dx = min(ra[2], rb[2]) - max(ra[0], rb[0])
                    dy = min(ra[3], rb[3]) - max(ra[1], rb[1])
                    SI = max(0, dx) * max(0, dy)
                    SU = SA + SB - SI
                    overlap_AB = float(SI) / float(SU)
                    if overlap_AB >= 0.15: # overlapping
                        rpn_List.append(rb)
                    b=b+1
                
            if len(rpn_List)>0:
                
                List.append(rpn_List)
            a = a + 1

        #print(List)
        F_Boxes = []
        for overalap_rpn in List:
            overalap_rpn = np.array(overalap_rpn)
            
            #percentage=len(overalap_rpn)/len(all_rpns)#need to check
            percentage=len(overalap_rpn)/16.0
            if percentage >= 0.4: # consistency 60 %
                #print("------------------")
                #print(overalap_rpn)
                
                arr = np.array(overalap_rpn)
                #print(arr.min(axis=0))
                #print("------------------")
                boxMin=arr.min(axis=0)
                #box1=overalap_rpn[0]
                boxMax=arr.max(axis=0)
                box1=[boxMin[0],boxMin[1],boxMax[2],boxMax[3]]
                '''box = [min_val[0], min_val[1], max_val[2], max_val[3]]'''
                if (box1[0]-10)>=0:
                    box1[0]= box1[0]-10

                if (box1[2] + 10) < 320:
                    box1[2] = box1[2] + 10

                if (box1[1] - 10) >= 0:
                    box1[1] = box1[1] -10

                if (box1[3] + 10) < 240:
                    box1[3] = box1[3] + 10
                box = [box1[0], box1[1], box1[2], box1[3]]
                #box = [box1[0], box1[1], box1[0]+box1[2], box1[1]+box1[3]]
                F_Boxes.append(box)
            else:
                print("Not Overlapped")
        #print(len(F_Boxes))
        return F_Boxes
```

```python
def get_all_rpns(vid):
        #print("label")
        vid_label=(vid.split('/')[-1]).split('.')[0]
        vid = (vid.split('/')[-1]).split('.')[0]
        rpns = []
        for rpn_line in rpn_lines:
            cleanedLine = rpn_line.strip()
            if cleanedLine:
                if vid in cleanedLine:
                    #print(cleanedLine)
                    rpn = cleanedLine.split(":")

                    #sort according to frame number
                    if rpn[1] == 'person':

                        frame_num=(rpn[0].split('.')[0]).split('_')[-1]
                        box = [int(float(rpn[2])), int(float(rpn[3])), int(float(rpn[4])+float(rpn[2])), int(float(rpn[5])+float(rpn[3])),int(frame_num)]

                        rpns.append(box)

        
        rpns=sorted(rpns, key=lambda x: x[4])
        #print(rpns)
        final_boxes = get_rpn(rpns,vid_label)
        #print(final_boxes)
        return final_boxes
```

```python
for folder in os.listdir(input_path):
    
    '''Will iterae for all folders containing videos or all categories of videos in data set'''
    if folder is not '.ipynb_checkpoints':
        fullPath = os.path.join(input_path, folder)
        if os.path.isdir(fullPath):
            for files in os.listdir(fullPath):
                '''Will iterate for every video in every folder or category in data set'''
                video=os.path.join(fullPath,files)
                
                print(video)
                rois=get_all_rpns(video)
                #print(rois)
                
                #need to consider videos which dont have detected ROI
                roi_count=0
                for roi in rois:
                    print(roi)
                    
                    #top, right, bottom, left = 10, 450+10, 360+10, 10  # Sample values.
                    #top, right, bottom, left = roi[0],roi[1],roi[2],roi[3]
                    left, bottom, right, top  = roi[0],roi[1],roi[2],roi[3]
                    '''print(left)
                    print(bottom)
                    print(right)
                    print(top)'''
                    #left, bottom, right, top  = 0,50,50,0
                    input_video = cv2.VideoCapture(video)
                    
                    #frame_width = int( input_video.get(cv2.CAP_PROP_FRAME_WIDTH))
                    #frame_height =int( input_video.get( cv2.CAP_PROP_FRAME_HEIGHT))
                    
                    frame_width=roi[2]-roi[0]#r-l
                    frame_height=roi[3]-roi[1]#t-b
                    
                    print(frame_height)
                    print(frame_width)
                    

                    fourcc = cv2.VideoWriter_fourcc(*'MJPG')
                    #fourcc = cv2.VideoWriter_fourcc(*'DIVX')
                    out_vid='/home/my_model/my_data/depth_'+str(depth)+'/'+str(interval)+'/'+str(roi_count)+'_'+video.split('/')[-1]
                    #output_movie = cv2.VideoWriter(out_vid, fourcc, 30, (frame_height, frame_width))
                    output_movie = cv2.VideoWriter(out_vid, fourcc, 30, (frame_width, frame_height))
                    
                    while True:
                        ret, frame = input_video.read()
                        
                        if not ret:
                            break

                            

                        # Following crop assumes the video is colored, 
                        # in c)ase it's Grayscale, you may use: crop_img = frame[top:bottom, left:right]  
                        #crop_img = frame[0:240, 0:320, :]
                        crop_img = frame[bottom:top, left:right, :]
    
                        #print(crop_img.shape)
                        output_movie.write(crop_img)


                    # Closes the video writer.
                    input_video.release()
                    output_movie.release()
                    
                    roi_count+=1

                    #totalFrames=cap.extract_frames(video,depth,interval) #-----------Zainab model
                    print("-------------Extracted ROI for "+video+"--------------")
```
