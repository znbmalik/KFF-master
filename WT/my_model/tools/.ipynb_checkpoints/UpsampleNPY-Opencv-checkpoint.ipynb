```python
def nn_interpolate(A, new_size,indx):
    
    """Vectorized Nearest Neighbor Interpolation"""
    old_size = A.shape # [H] x [W] 35x25
    #print(old_size)
    
    row_ratio,col_ratio = np.array(new_size)/np.array(old_size)
    '''print("---------------")
    print(row_ratio)
    print(col_ratio)
    print("---------------")
    #row_ratio=1.66
    #col_ratio=2.33
    #col_ratio=2.32'''
    
    # column wise interpolation
    col_idx = (np.floor(range(1, 1 + int(old_size[1]*col_ratio))/col_ratio) - 1).astype(int)
    
    # row wise interpolation 
    row_idx = (np.floor(range(1, 1 + int(old_size[0]*row_ratio))/row_ratio) - 1).astype(int)
   
   
    #print(row_idx)
    #print(col_idx)
    #final_matrix = A[:, row_idx][col_idx, :]
    final_matrix = A[row_idx,:][:,col_idx]
    #print(final_matrix.shape)
    if indx%2 == 0:
        final_matrix*=row_ratio
    else:
        final_matrix*=col_ratio
    return final_matrix
```
